Preprocessing

```python

df['discounted_price'] = df['discounted_price'].replace('[₹,]', '', regex=True).astype(float)
df['actual_price'] = df['actual_price'].replace('[₹,]', '', regex=True).astype(float)

df['discount_percentage'] = df['discount_percentage'].astype(str).str.replace('%', '').astype(float)
df['rating_count'] = df['rating_count'].replace(',', '', regex=True).fillna(0).astype(int)

```

Cleaning

```python

df['category'] = df['category'].str.strip().str.title()
df['product_name'] = df['product_name'].str.strip()
```

Visualizations
1. Bar Chart – Top 10 Categories by Total Discounted Sales

```python
top10_category_sales = df.groupby('category')['discounted_price'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=top10_category_sales.index, y=top10_category_sales.values, color='orange')
plt.title('Top 10 Categories by Total Discounted Sales')
plt.xlabel('Product Category')
plt.ylabel('Total Discounted Sales (₹)')
plt.xticks(rotation=75, ha='right')
plt.tight_layout()
plt.show()
```

 2. Pie Chart – Product Distribution by Category (Top 6 Categories)

```python
category_counts = df['category'].value_counts().head(6)  # Top 6 for clarity
colors = ['lightcoral', 'gold', 'skyblue', 'lightgreen', 'mediumpurple', 'orange']

plt.figure(figsize=(6,6))
category_counts.plot(kind='pie', autopct='%1.1f%%', colors=colors)
plt.title("Product Distribution by Category")
plt.ylabel("")  # Hide y-axis
plt.tight_layout()
plt.show()
```

3. Heatmap – Correlation Between Key Numeric Features

```python
# Select relevant numeric columns
numeric_df = df[['discounted_price', 'actual_price', 'discount_percentage', 'rating_count']]

plt.figure(figsize=(8,6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title("Correlation Matrix of Amazon Product Features")
plt.tight_layout()
plt.show()
```
